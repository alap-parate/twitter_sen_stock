# Extract Tweets

```python
string ="#nifty since:" + str(yesterday)+ " until:" + str(today) +"-is:retweet"
# string = ['nse','nifty'] 
# tweets_list= tweepy.Cursor(api.search_tweets, tweet_mode='extended',
#                            q=string, lang='en').items()

tweets_list = tweepy.Cursor(api.search_tweets, q=string, lang='en').items()

output = []
for tweet in tweets_list:
    text = tweet._json["text"]
    print(text)
    favourite_count = tweet.favorite_count
    retweet_count = tweet.retweet_count
    created_at = tweet.created_at
    line = {'text' : text, 'favourite_count' : favourite_count,'retweet_count' : retweet_count,
           'created_at' : created_at}
    output.append(line)
```

```python
df = pd.DataFrame(output)
df
```
